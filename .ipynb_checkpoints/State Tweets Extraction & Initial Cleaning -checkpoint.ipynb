```python
from config import api_key, api_secret, bearer_token, access_token, access_secret
import tweepy
import time
#This sets the authorization for tweepy, seems to work
#Have to set auth every time. 
auth = tweepy.OAuthHandler(api_key, api_secret)
auth.set_access_token(access_token, access_secret)
api = tweepy.API(auth)
```

```python
#List of states to build tweet dictionary
states = ["Alabama","Alaska","Arizona","Arkansas","California","Colorado",
  "Connecticut","Delaware","Florida","Georgia","Hawaii","Idaho","Illinois",
  "Indiana","Iowa","Kansas","Kentucky","Louisiana","Maine","Maryland",
  "Massachusetts","Michigan","Minnesota","Mississippi","Missouri","Montana",
  "Nebraska","Nevada","New Hampshire","New Jersey","New Mexico","New York",
  "North Carolina","North Dakota","Ohio","Oklahoma","Oregon","Pennsylvania",
  "Rhode Island","South Carolina","South Dakota","Tennessee","Texas","Utah",
  "Vermont","Virginia","Washington","West Virginia","Wisconsin","Wyoming"]
state_tweets = dict()
```

```python
#Try building state dictionary here
loop = 0
for i in states:
    for status in tweepy.Cursor(api.search, q = f'public transit {i} exclude:retweets', tweet_mode = 'extended').items(100000):
        #If state already in dict, append additional tweet
        if i in state_tweets:
            state_tweets[i].append(status.full_text)
        #If not, add state to dict and tweet.
        else:
            state_tweets[i] = [status.full_text]
        time.sleep(1.5)
            
        loop +=1
        print(f"Loop {loop} complete.")
        print(f"State {i} complete.")
print(state_tweets)
```

```python
#Make everything one big string
def combine_text(list_of_text):
    '''Takes a list of text and combines them into one large chunk of text.'''
    combined_text = ' '.join(list_of_text)
    return combined_text
```

#### Now, let's work on cleaning data. 

```python
#Ex of what a sample tweet text looks like.
test = state_tweets[0].loc[3]
test
test2 = re.sub('@[^\s]+','',test)
test2
```

```python
#Now, some potential regex for cleaning this data. 
def clean_text_round1(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('@[^\s]+','',text)
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

round1 = lambda x: clean_text_round1(x)
```

```python
state_tweets['tweets'] = state_tweets['tweets'].apply(round1)
state_tweets.tweets[0]
```

### Remove stop words/other preprocessing

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
nltk.download('stopwords')
nltk.download('punkt')
```

```python
#Tokenize df
state_tweets['tweets'] = state_tweets['tweets'].apply(word_tokenize)
state_tweets.head()
```

```python
#Take out stop words 
stop_words = set(stopwords.words('english')) 
stop = lambda x: [w for w in x if not w in stop_words] 
state_tweets['tweets'] = state_tweets['tweets'].apply(stop)
state_tweets.head()
#In this column, state_tweets.head() now has the stop words removed. 
```

```python
#Pickle corpus just in case. 
import pickle
```

Use CountVectorizer to take out stop words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words = 'english')
tweets_cv = cv.fit_transform(state_tweets.tweets)
tweets_cv
```

```python
tweets_dtm = pd.DataFrame(tweets_cv.toarray(), columns = cv.get_feature_names())
tweets_dtm.index = state_tweets.state
tweets_dtm
```

```python
#Export to csv
#Export as CSV 
state_tweets.to_csv('state_tweets_corpus.csv')
```
