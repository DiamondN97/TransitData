```python
#Initial Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Bring in sentiment data
sentiment = pd.read_csv('sentiment.csv')
sentiment = sentiment[['state', 'polarity', 'subjectivity']]
sentiment
```

```python
#Bring in fares data
fares = pd.read_csv('fares.csv')
fares.head()
```

```python
#Revise columns
fares.columns = ['state', 'fares']
fares.head()
```

```python
#Join sentiment and fares table
sentiment_fares = pd.merge(sentiment, fares, how='inner', on=['state'])
sentiment_fares
```

```python
#Bring in funding and ridership data 
funding_ridership = pd.read_csv('funding_ridership.csv')
funding_ridership.columns = ['year', 'urban_funding', 'total_ridership']
funding_ridership.head()
```

```python
#Scatterplot for funding vs ridership
plt.scatter(funding_ridership.urban_funding, funding_ridership.total_ridership)
plt.xlabel('Urban Funding for Transit (in millions)')
plt.ylabel('Ridership of Public Transit (in millions)')
```

```python
funding_ridership = funding_ridership.dropna(axis = 0)
funding_ridership
```

```python
# feature engineering - 
# add 2 lags previous year and 2 years ago-ridership
prev_year = funding_ridership.total_ridership[:-1]
two_yr = funding_ridership.total_ridership[:-2]
two_yr
```

```python
concat = pd.concat([funding_ridership, prev_year], axis = 1)
# test_con['gdp'] = df['gdp'].shift(-1)
concat.columns = ['year', 'urban_funding', 'total_ridership', 'last_years_ridership']
concat['last_years_ridership'] = concat['last_years_ridership'].shift(1)
concat = concat.replace(np.nan, '', regex=True)
funding_ridership_feature_engineer = pd.concat([concat, two_yr], axis = 1)
funding_ridership_feature_engineer.columns = ['year', 'urban_funding', \
                                             'total_ridership', 'last_years_ridership',\
                                             'two_yrs_ago_ridership']
funding_ridership_feature_engineer['two_yrs_ago_ridership'] = funding_ridership_feature_engineer['two_yrs_ago_ridership'].shift(2)
funding_ridership_feature_engineer = funding_ridership_feature_engineer.replace(np.nan, '', regex=True)
funding_ridership_feature_engineer
```

```python
#Drop first two rows for this pass
lagdf = funding_ridership_feature_engineer.drop(funding_ridership_feature_engineer.index[0:2])
lagdf
```

```python
#Build regression for funding vs ridership
X = lagdf[['urban_funding', 'last_years_ridership', 'two_yrs_ago_ridership']].values.reshape(-1, 3)
y = lagdf['total_ridership'].values.reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
scaler_y=StandardScaler()
scaler_y.fit(y_train)
y_train_scaled=scaler_y.transform(y_train)
y_test_scaled=scaler_y.transform(y_test)
```

```python
#Now, set up training the model on training data
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train_scaled, y_train_scaled)
```

```python
#Now make predictions!
predictions = lm.predict(X_test_scaled)
```

```python
#Make scatter plot of y_test values vs predictions to see if the model is any good
plt.scatter(y_test_scaled, predictions)
plt.xlabel("y_test (known values)")
plt.ylabel("model predictions")
```

```python
from sklearn import metrics
MAE = metrics.mean_absolute_error(y_test_scaled, predictions)
MSE = metrics.mean_squared_error(y_test_scaled, predictions)
RMSE = np.sqrt(MSE)
```

```python
#Print final results
print(f'MAE: {MAE}')
print(f'MSE: {MSE}')
print(f'RMSE: {RMSE}')
#Obvi these seem high, but within about 10% seems ok? 
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(x=X[:,0], y=scaler_y.transform(y))
plt.scatter(x=X[:,0], y=lm.predict(scaler.transform(X)))
plt.xlabel('Urban Public Transit Funding ($ millions)')
plt.ylabel('Transit Ridership ($ millions)')
plt.title('Scikit-Learn Linear Regression Model', fontsize = 15)
plt.savefig('mlmodel.png')
```

```python
print(lm.coef_)
print(lm.intercept_)
```

```python
import seaborn as sns
sns.distplot((y_test-predictions))
```
